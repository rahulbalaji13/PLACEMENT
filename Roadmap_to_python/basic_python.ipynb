# 5. Inheritance



````python
class Demo:
    def __init__(self):
        self.x = 1
    def change(self):
        self.x = 10  # Not executed because --> When obj.change() is called,
        '''
        Python looks for the change method in the Demo_derived class first.
        Since Demo_derived does have its own change method, that method is executed,
        and the change method from the Demo class is not called.

        '''
class Demo_derived(Demo):
    def change(self):
        self.x=self.x+1
        return self.x
def main():
    obj = Demo_derived()
    print(obj.change())

main()
````

Let's trace the execution of the code:

1.  **`main()` function call**: The program begins by calling the `main()` function.

2.  **`obj = Demo_derived()`**: An object `obj` is created from the `Demo_derived` class.
    *   When `Demo_derived` is instantiated, its constructor (if present) would be called. Since `Demo_derived` doesn't have its own `__init__` method, Python calls the `__init__` method of its parent class, `Demo`.
    *   Inside `Demo.__init__(self)`, `self.x` is initialized to `1`. So, `obj.x` becomes `1`.

3.  **`print(obj.change())`**: The `change()` method is called on the `obj` object.
    *   Since `Demo_derived` has its own `change()` method, that method is executed (this is method overriding).
    *   Inside `Demo_derived.change(self)`:
        *   `self.x` (which is `1`) is updated by `self.x = self.x + 1`.
        *   So, `self.x` becomes `1 + 1 = 2`.
        *   The method then `return self.x`, which is `2`.

4.  **`print()` output**: The `print()` function receives the value `2` from `obj.change()` and outputs it to the console.

Therefore, the final output of the code is:
```
2
```

 # 8. Multithreading

````python
import threading

class thread(threading.Thread):
    def __init__(self, thread_ID, thread_name):
        threading.Thread.__init__(self)
        self.thread_ID = thread_ID
        self.thread_name = thread_name
    def run(self):
        print(self.thread_name)

thread1 = thread(100, "Student")
thread2 = thread(101, "VJIT ")
thread3 = thread(102, "VJITStudent ")

thread = []
thread.append(thread1)
thread.append(thread2)
thread.append(thread3)

thread1.start()
thread2.start()
for thread in thread:
    thread.join()

thread3.start()

print("Exit")
````

## Right-Angled Triangle Pattern



````python
print("## Right-Angled Triangle Pattern\n\n### Subtask:\nGenerate Python code to print a right-angled triangle pattern using asterisks.\n")
# Print a right-angled triangle pattern using asterisks based on a user-defined size.

# Prompt the user to enter the desired size (number of rows) for the triangle.
size = int(input("Enter the size of the right-angled triangle (e.g., 5 for 5 rows): "))

# Outer loop for iterating through each row.
# The range(size) function generates numbers from 0 up to (but not including) 'size'.
# We will use 'i' to represent the current row index, starting from 0.
for i in range(size):
    # Inner loop for iterating through each column in the current row.
    # The number of asterisks in each row should be (i + 1) to form the triangle.
    for j in range(i + 1):
        # Print an asterisk followed by a space. The 'end=" "' argument prevents a newline
        # after each asterisk, keeping them on the same line.
        print("*", end=" ")
    # After the inner loop completes (i.e., all asterisks for the current row are printed),
    # print a newline character to move to the next line for the next row.
    print()
````

## Pyramid Pattern



````python
print("## Pyramid Pattern\n\n### Subtask:\nGenerate Python code to print a pyramid pattern (equilateral triangle) using asterisks.\n")
# Print a pyramid pattern using asterisks based on a user-defined size.

# Prompt the user to enter the desired size (number of rows) for the pyramid.
size = int(input("Enter the size of the pyramid (e.g., 5 for 5 rows): "))

# Outer loop for iterating through each row.
# The range(size) function generates numbers from 0 up to (but not including) 'size'.
# 'i' represents the current row index.
for i in range(size):
    # Inner loop for printing leading spaces.
    # The number of spaces decreases with each row: (size - i - 1).
    for j in range(size - i - 1):
        print(" ", end="") # Print a space

    # Inner loop for printing asterisks.
    # The number of asterisks increases with each row: (2 * i + 1).
    for k in range(2 * i + 1):
        print("*", end="") # Print an asterisk

    # After all spaces and asterisks for the current row are printed,
    # print a newline character to move to the next line.
    print()
````

## Diamond Pattern



````python
print("## Diamond Pattern\n\n### Subtask:\nGenerate Python code to print a diamond pattern using asterisks.\n")
# Print a diamond pattern using asterisks based on a user-defined size.

# Prompt the user to enter the desired size (number of rows for the upper half) for the diamond.
size = int(input("Enter the size for the diamond (e.g., 5 for 5 rows in the upper half): "))

# --- Upper half of the diamond (Pyramid) ---
# Outer loop for iterating through each row of the upper half.
# 'i' represents the current row index, from 0 to size-1.
for i in range(size):
    # Inner loop for printing leading spaces.
    # The number of spaces decreases with each row: (size - i - 1).
    for j in range(size - i - 1):
        print(" ", end="") # Print a space

    # Inner loop for printing asterisks.
    # The number of asterisks increases with each row: (2 * i + 1).
    for k in range(2 * i + 1):
        print("*", end="") # Print an asterisk

    # After all spaces and asterisks for the current row are printed,
    # print a newline character to move to the next line.
    print()

# --- Lower half of the diamond (Inverted Pyramid) ---
# Outer loop for iterating through each row of the lower half.
# We start from 'size - 2' (to avoid duplicating the widest row) down to 0.
# 'i' represents the current row index for the inverted pyramid.
for i in range(size - 2, -1, -1):
    # Inner loop for printing leading spaces.
    # The number of spaces increases with each row: (size - i - 1).
    for j in range(size - i - 1):
        print(" ", end="") # Print a space

    # Inner loop for printing asterisks.
    # The number of asterisks decreases with each row: (2 * i + 1).
    for k in range(2 * i + 1):
        print("*", end="") # Print an asterisk

    # After all spaces and asterisks for the current row are printed,
    # print a newline character to move to the next line.
    print()
````

## Number Pattern



````python
print("## Number Pattern\n\n### Subtask:\nGenerate Python code to print a number pattern where each row displays sequential numbers.\n")
# Print a number pattern where each row displays sequential numbers based on a user-defined size.

# Prompt the user to enter the desired number of rows for the pattern.
size = int(input("Enter the number of rows for the number pattern (e.g., 5): "))

# Initialize a counter for sequential numbers.
num = 1

# Outer loop for iterating through each row.
# 'i' represents the current row index, from 0 to size-1.
for i in range(size):
    # Inner loop for iterating through each column in the current row.
    # The number of elements in each row increases with each row (i+1).
    for j in range(i + 1):
        # Print the current sequential number, followed by a space.
        # The 'end=" "' argument prevents a newline after each number, keeping them on the same line.
        print(num, end=" ")
        # Increment the counter for the next sequential number.
        num += 1
    # After the inner loop completes (i.e., all numbers for the current row are printed),
    # print a newline character to move to the next line for the next row.
    print()
````

# 11. Heap queue or heapq data structure in python

A heap queue (also called priority queue) is a special data structure that allow quick access to the smallest (min-heap) or largest (max-heap). min-heap was implemented in default. Smallest element will be always at root of the strucutre for offering effient access.

NOTE: Mostly heapq was used in dijikstra algm, huffman coding or any task that require quick access to smallest element

Helps to maintain a list in heap order with minimal code and high performance

They are ideal for managing sorted data dynamically without full sorting after each operation


## Key operations:

1. Create (heapify): Convert a regular list into a valid min-heap using heapq.heapify()

2. Push (heappush): Adds an element to the heap while keeping the heap properly intact

3. Pop (heappop): Removes and returns the smallest element from the heap

4. Peek: Access the smallest element without removing it using heap[0]

5. Push and pop (heappushpop): Push a new element and pop the smallest in a single step

6. Replace(heapreplace): Pop the smallest and push a new element in one step

````python
# Default min heap
import heapq
lst = [10,20,30,40,50]

heapq.heapify(lst)
print("Min Heap Queue:", lst)

# Since heapq is implementing min heap in default,
#To create a max-heap,
#we need to invert the values(store negative number)
nums = [30,40,50,60,70]
max_heap = [-n for n in nums] # Convert into max heap by inverting values
heapq.heapify(max_heap)
print("Print the largest element (used max heap):", -max_heap[0]) # Access the largest element (invert sign again)
````

````python
# TO push element and pop the smallest element
heapq.heappush(lst, 60)
print(lst)
heapq.heappop(lst)
print(lst) # THis pops smallest element
````

````python
#Push and pop simultaneously
min = heapq.heappushpop(lst, 70)
print(min)
print(lst)
````

````python
# Find the largest and smallest element upto specific range - Use nlargest() and nsmallest()
large = heapq.nsmallest(3, lst)
small = heapq.nlargest(3, lst)
print(large)
print(small)
````

````python
# Preplace and merge operations
lst1 = [40,50,60,70]
print(lst1)
#Replacing smallest element 40 with 30
min_elem_replaced = heapq.heapreplace(lst1, 30)
print(min_elem_replaced)
print(lst1)

lst2 = [40, 80, 90, 100]
merge = list(heapq.merge(lst1, lst2))
print("Merged heap:", merge)
````

# 12. Deque - Double ended queue

Deque stands for double ended queue. It allows us to add and remove elements from both side efficiently.

Applications: Task sheduling, Sliding window problems and real-time data processing.

It offers built-in methods : appendleft(), popleft() and rotate()

It can function both as queue and stack

Takes O(1) for adding/removing elements

````python
from collections import deque
import collections

q = collections.deque([1,2,3]) # Declaring queue

# Append(x): Adds x to the right end of the deque.
q.append(4)
print("After append right 4:", q)
# Eppendleft(x): Adds x to the left end of the deque.
q.appendleft(5)
print("After append left 5:", q)
# extend(iterable): Adds all elements from the iterable to the right end.
q.extend([4,5,6])
print("After extend([4,5,6]):", q)

# Extendleft(iterable): Add all elements from the iterable to the left end (in reverse order)
q.extendleft([0])
print("After extendleft([0]):", q)

# Remove method
q.remove(6)
print("After remove(6):", q)

# Remove elements from th right
q.pop()

# Remove elements from the left
q.popleft()

print("After pop and popleft:", q)

#Finding length of queue
print(len(q))

# Counting the occurence of the value
print(q.count(5))

# Rotating the deque for the 2 steps to the right
q.rotate(2)
print(q)

# Rotating the deque for the 3 steps to the left
q.rotate(-3)
print(q)

# Reversing the deque
q.reverse()
print(q)

# Accessing element by index:
print(q[-1])
print(q[0])

# Clear all elem in queue
q.clear()

print("After clear():", q)
````

# 13. Tuples and Dictionary

'''
Comparison: Tuples vs. Dictionaries

Tuples:
- Immutable: Once created, elements cannot be changed, added, or removed.
- Ordered: Elements have a defined order, which will not change.
- Allow duplicates: Can contain multiple items with the same value.
- Accessed by index: Elements are retrieved using integer indices (e.g., my_tuple[0]).
- Defined using parentheses: my_tuple = (1, 2, 3).
- Use cases: Storing collections of related but fixed data (e.g., coordinates, database records).

Dictionaries:
- Mutable: Elements can be changed, added, or removed after creation.
- Ordered (from Python 3.7+): Insertion order is preserved. (Prior to 3.7, they were unordered).
- Key-value pairs: Stores data in key: value pairs.
- Keys must be unique and immutable: Each key must be distinct and of an immutable type (e.g., strings, numbers, tuples).
- Accessed by key: Values are retrieved using their associated keys (e.g., my_dict["key"]).
- Defined using curly braces: my_dict = {"key": "value"}.
- Use cases: Storing data where each piece of information has a unique identifier (e.g., user profiles, configuration settings).
'''

````python
# --- Tuple Methods ---
print("\n--- Tuple Methods ---")
my_tuple = (1, 2, 3, 2, 4, 5, 2)
print("Original Tuple:", my_tuple)

# 1. count(): Returns the number of times a specified value occurs in a tuple
print("Count of 2:", my_tuple.count(2)) # Output: 3

# 2. index(): Searches the tuple for a specified value and returns the position of where it was found
print("Index of 4:", my_tuple.index(4)) # Output: 4

# Tuples also support slicing and indexing (not methods, but common operations)
print("Element at index 0:", my_tuple[0]) # Output: 1
print("Slice from index 1 to 4:", my_tuple[1:5]) # Output: (2, 3, 2, 4)


# --- Dictionary Methods ---
print("\n--- Dictionary Methods ---")
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
print("Original Dictionary:", my_dict)

# 1. clear(): Removes all the elements from the dictionary
# temp_dict = my_dict.copy() # Make a copy to demonstrate clear without affecting original for other tests
# temp_dict.clear()
# print("After clear():", temp_dict) # Output: {}

# 2. copy(): Returns a copy of the dictionary
new_dict = my_dict.copy()
print("Copied Dictionary:", new_dict)

# 3. fromkeys(): Returns a dictionary with the specified keys and value
keys = ('key1', 'key2', 'key3')
value = 0
dict_from_keys = dict.fromkeys(keys, value)
print("fromkeys():", dict_from_keys) # Output: {'key1': 0, 'key2': 0, 'key3': 0}

# 4. get(): Returns the value of the specified key
print("Get 'name':", my_dict.get("name")) # Output: Alice
print("Get non-existent key (returns None):"), my_dict.get("country") # Output: None
print("Get non-existent key with default value:", my_dict.get("country", "USA")) # Output: USA

# 5. items(): Returns a list containing a tuple for each key-value pair
print("items():", my_dict.items())

# 6. keys(): Returns a list containing the dictionary's keys
print("keys():", my_dict.keys())

# 7. pop(): Removes the element with the specified key
popped_age = my_dict.pop("age")
print("After pop('age'):", my_dict, "Popped value:", popped_age)

# 8. popitem(): Removes the last inserted key-value pair (in versions before 3.7, a random item is removed)
popped_item = my_dict.popitem()
print("After popitem():", my_dict, "Popped item:", popped_item)

# Reset my_dict for further demonstrations
my_dict = {"name": "Alice", "age": 30, "city": "New York"}

# 9. setdefault(): Returns the value of the specified key. If the key does not exist: inserts the key with the specified value.
my_dict.setdefault("occupation", "Engineer")
print("After setdefault('occupation'):", my_dict)
my_dict.setdefault("name", "Bob") # Key already exists, value not changed
print("After setdefault('name'):", my_dict)

# 10. update(): Updates the dictionary with the specified key-value pairs
my_dict.update({"age": 31, "country": "USA"})
print("After update():", my_dict)

# 11. values(): Returns a list of all the values in the dictionary
print("values():", my_dict.values())

# Other common dictionary operations (not methods):
# Accessing a value by key
print("Direct access 'name':", my_dict["name"])

# Adding a new key-value pair
my_dict["email"] = "alice@example.com"
print("After adding 'email':", my_dict)

# Modifying an existing value
my_dict["age"] = 32
print("After modifying 'age':", my_dict)

# Deleting a key-value pair using del keyword
del my_dict["city"]
print("After deleting 'city':", my_dict)
````
