# 1. SETS

**SET: Sets are used to store multiple items in a single variable that are unordered, unchangeable and do not allow duplicate value. SET = {}**

````python
mysets = {1,2,3}

print(mysets)

print(len(mysets))

mysets.add(4)

print(mysets)

mysets1 = {5,6,7}

mysets.update(mysets1)

print(mysets)

mysets.remove(3)

print(mysets)

mysets.discard(2)

print(mysets)

mysets.clear()

print(mysets)
````

# 2. Dictionary

**Dictionaries are used to store data values in keywords and value pairs**


````python
mydict = {
    "IIT" : "Indian Institute of Technology",
    "ISRO" : "Indian Space Research Organisation",
    "TCS" : "Tata Consultancy Services"
}

print(mydict)

print(mydict["TCS"])

mydict["IIT"] = "Indian Institute of Technology Madras"

print(mydict)

print(mydict.keys())

print(mydict.values())

print(mydict.items())
````

````python
class class_name:
  def __init__(self, name, brand, engine, mileage):
    self.name = name
    self.brand = brand
    self.engine = engine
    self.mileage = mileage

object_name_car = class_name("M4", "BMW", 4300, 10)
print(object_name_car.brand)
````

````python
class fruits:
    def __init__(self, price):
        self.price = price
obj=fruits(50)

obj.quantity=10
obj.bags=2

# obj.quantity is 10
# obj.__dict__ contains {'price': 50, 'quantity': 10, 'bags': 2}
# len(obj.__dict__) will be 3 (number of attributes: price, quantity, bags)
# So, 10 (obj.quantity) + 3 (len(obj.__dict__)) = 13
print(obj.quantity+len(obj.__dict__))
````

````python
class Employee:
    def __init__(self):
        self.id=1000                       #PUBLIC ACCESS MODIFIER
        self._salary=10000                 #PRIVATE ACCESS MODIFIER
emp=Employee()
print(emp.id)
print(emp.salary)  #SINCE SALARY IS DECLARED WITH _SALARY, IT CAN'T BE ACCESSED
````

````python
class python_string:
    def reverse_words(self, sentence):
        words = sentence.split()
        reversed_words = words[::-1]
        return ' '.join(reversed_words)

# Sample usage:
string_reverser = python_string()
input_sentence = "Hello Python"
output_sentence = string_reverser.reverse_words(input_sentence)
print(output_sentence)

input_sentence_2 = "This is a test sentence"
output_sentence_2 = string_reverser.reverse_words(input_sentence_2)
print(output_sentence_2)
````

````python
class Circle():
    def __init__(self, r):
        self.radius = r

    def area(self):
        return self.radius**2*3.14

    def perimeter(self):
        return 2*self.radius*3.14

radius = int(input())
NewCircle = Circle(radius)
print(NewCircle.area())
print(NewCircle.perimeter())
````

````python
class IOString:
    def __init__(self,s):
        self.s = s

    def get_String(self):
        self.s = input()

    def print_String(self):
        return self.s.upper()


myobject = IOString('')
myobject.get_String()
print(myobject.print_String())
````
